```python
import jaro
from collections import defaultdict
entity_string_1 = "My giant test"
entity_string_2 = "Test my big giant test "
def to_n_grams(s, n =2):
    s = " "+s+" "
    s_gram = defaultdict(lambda : 0)
    for i in range(len(s)-n +1) :
        s_gram[s[i:i+2]] = s_gram[s[i:i+2]] +1
    return s_gram    

def q_similarity(s,t):
    s_gram = to_n_grams(s)
    t_gram = to_n_grams(t)
    
    all_gram = s_gram + t_gram
    
def similarity(entity_string_1, entity_string_2) :
    print()
    print(f"Tested values : {entity_string_1}, {entity_string_2}")
    print(f"Direct Jaro-Winkler score : {round(jaro.jaro_winkler_metric(entity_string_1, entity_string_2),3)}")

    entity_string_1_splitted = entity_string_1.split()
    entity_string_2_splitted = entity_string_2.split()
    #test with a circular loop of the splitted entity
    scores = []
    for i in range(len(entity_string_1_splitted)):
        e_1 = ' '.join(entity_string_1_splitted[-i:] + entity_string_1_splitted[:-i])
        for i in range(len(entity_string_2_splitted)):
            e_2 = ' '.join(entity_string_2_splitted[-i:] + entity_string_2_splitted[:-i])
            scores.append(round(jaro.jaro_winkler_metric(e_1, e_2),3))
    print(f"Looping through the string and use Jaro-Winkler score : {max(scores)}")
    
    scores = []
    for i in range(len(entity_string_1_splitted)):
        e_1 = ' '.join(entity_string_1_splitted[-i:] + entity_string_1_splitted[:-i])
        for i in range(len(entity_string_2_splitted)):
            e_2 = ' '.join(entity_string_2_splitted[-i:] + entity_string_2_splitted[:-i])
            scores.append(round(jaro.jaro_metric(e_1, e_2),3))
    print(f"Looping through the string and use Jaro score : {max(scores)}")
    
    scores = []
    for i in range(len(entity_string_1_splitted)):
        e_1 = ' '.join(entity_string_1_splitted[-i:] + entity_string_1_splitted[:-i])
        for i in range(len(entity_string_2_splitted)):
            e_2 = ' '.join(entity_string_2_splitted[-i:] + entity_string_2_splitted[:-i])
            scores.append(round(jaro.jaro_metric(e_1, e_2),3))
    print(f"Looping through the string and use Jaro score : {max(scores)}")
    print()
    
similarity("My giant test","Test my big giant test")
similarity("Marco Polo","Polo Marco")                 # <-- This case is the most important and the difference is huge
similarity("Marco Polo","Jean Pierre")
similarity("Marco Polo","Un ami de Marco Polo")       # <-- This case is also interesting
similarity("Marco Polo","Marco Jean jaques")       
similarity("Marco Polo","Polo")    
similarity("Marco Polo","Marco")   
```
